```python
df_claims_by_county = pd.read_sql("SELECT * FROM marts.claims_by_county", engine)
df_claims_by_month = pd.read_sql("SELECT * FROM marts.claims_by_month", engine)
df_loss_ratio_by_month = pd.read_sql("SELECT * FROM marts.loss_ratio_by_month", engine)

print(df_claims_by_county.head())
print(df_claims_by_month.head())
print(df_loss_ratio_by_month.head())
# Visualize claims by county
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
df_claims_by_county.sort_values('claims_count', ascending=False).plot(
    x='county', y='claims_count', kind='bar', legend=False, ax=plt.gca()
)
plt.title('Claims Count by County')
plt.ylabel('Claims Count')
plt.xlabel('County')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Visualize claims by month
plt.figure(figsize=(12,6))
df_claims_by_month.plot(
    x='month', y='claims_count', kind='line', marker='o', legend=False, ax=plt.gca()
)
plt.title('Claims Count by Month')
plt.ylabel('Claims Count')
plt.xlabel('Month')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Visualize loss ratio by month
plt.figure(figsize=(12,6))
df_loss_ratio_by_month.plot(
    x='month', y='loss_ratio_pct', kind='line', marker='o', legend=False, ax=plt.gca()
)
plt.title('Loss Ratio (%) by Month')
plt.ylabel('Loss Ratio (%)')
plt.xlabel('Month')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns

import matplotlib.pyplot as plt

# Claims status distribution
df_claims['status'].value_counts().plot(kind='bar', title='Claims Status Distribution')
plt.ylabel('Count')
plt.tight_layout()
plt.show()

# Claims by product type
df_claims['product_type'].value_counts().plot(kind='bar', title='Claims by Product Type')
plt.ylabel('Count')
plt.tight_layout()
plt.show()

# Average paid amount by severity band
df_claims.groupby('severity_band')['paid'].mean().plot(kind='bar', title='Avg Paid by Severity Band')
plt.ylabel('Average Paid Amount')
plt.tight_layout()
plt.show()

# Top 10 counties by customer count
df_customers['county_name'].value_counts().head(10).plot(kind='bar', title='Top 10 Counties by Customer Count')
plt.ylabel('Customer Count')
plt.tight_layout()
plt.show()

# Average risk scores by county (heatmap)
risk_cols = ['crime_risk', 'hail_risk', 'flood_risk', 'wind_risk', 'fire_risk']
risk_by_county = df_customers.groupby('county_name')[risk_cols].mean().head(10)
plt.figure(figsize=(10,6))
sns.heatmap(risk_by_county, annot=True, cmap='YlOrRd')
plt.title('Average Risk Scores by County (Top 10)')
plt.tight_layout()
plt.show()

# Policy status distribution
df_policies['status'].value_counts().plot(kind='bar', title='Policy Status Distribution')
plt.ylabel('Count')
plt.tight_layout()
plt.show()

# Product type distribution in policies
df_policies['product_type'].value_counts().plot(kind='bar', title='Product Type Distribution in Policies')
plt.ylabel('Count')
plt.tight_layout()
plt.show()

# Average gross premium by channel
df_policies.groupby('channel')['gross_premium'].mean().plot(kind='bar', title='Avg Gross Premium by Channel')
plt.ylabel('Average Gross Premium')
plt.tight_layout()
plt.show()
```
